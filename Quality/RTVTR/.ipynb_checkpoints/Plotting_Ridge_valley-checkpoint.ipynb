```python
import dill
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import seaborn as sns
from sklearn import linear_model
from scipy import stats

Floc = '/home/citer/Desktop/GANS/Generated_images_512_BMP/RTVTR/Live_RTVTR_numpy/'
Floc2 = '/home/citer/Desktop/GANS/Generated_images_512_BMP/RTVTR/Generated_RTVTR_numpy/'
```

```python
'''Live Samples'''

files = os.listdir(Floc)
Live_RTVTR = []  #ridge to valley thinkens ratio
Live_RC =[]      # ridge Count
Live_WLC = []    #white line counts

for file in files:
    temp_clean = np.load(os.path.join(Floc,file), allow_pickle=True)
    temp_clean = temp_clean.item()
    '''Selecting blocks with best ridge valley uniformity'''
    d = temp_clean['Q_RVU_local']
    h = d.shape[0]
    w = d.shape[1]
    N_selectuion = 15        # selecting top 15 highest quality meaning lowest RVU
    x1 = np.reshape(d, -1)   #turn into 1d
    Num_nan = sum(sum(np.isnan(d)))   #count of nans
    Num_zeros = sum(x1 == 0)           #count of zeros
    xs = np.sort(x1)                  #sorted array
    s = np.argsort(x1)    #indecies sorted from low to high in original array 
    indices_s = s[Num_zeros : Num_zeros + N_selectuion]

    #cordinates
    x = [int(i/w) for i in indices_s]  
    y = indices_s%w
    Cordinates = list(zip(x,y))

    temp_RTVTR = []
    temp_RC = []
    temp_WLC = []

    for c in Cordinates:
        for r in range(len(temp_clean['Q_Ratios'][c])):
            temp_RTVTR.append(temp_clean['Q_Ratios'][c][0][r])
        temp_RC.append(temp_clean['Q_RC'][c])
        temp_WLC.append(temp_clean['Q_WLC'][c])

    Live_RTVTR.append(np.mean(temp_RTVTR))
    Live_RC.append(np.mean(temp_RC))
    Live_WLC.append(np.mean(temp_WLC))
```

```python
'''Generated Samples'''

files = os.listdir(Floc2)
Gen_RTVTR = []  #ridge to valley thinkens ratio
Gen_RC =[]      # ridge Count
Gen_WLC = []    #white line counts

for file in files:
    temp_clean = np.load(os.path.join(Floc2,file), allow_pickle=True)
    temp_clean = temp_clean.item()
    '''Selecting blocks with best ridge valley uniformity'''
    d = temp_clean['Q_RVU_local']
    h = d.shape[0]
    w = d.shape[1]
    N_selectuion = 15        # selecting top 15 highest quality meaning lowest RVU
    x1 = np.reshape(d, -1)   #turn into 1d
    Num_nan = sum(sum(np.isnan(d)))   #count of nans
    Num_zeros = sum(x1 == 0)           #count of zeros
    xs = np.sort(x1)                  #sorted array
    s = np.argsort(x1)    #indecies sorted from low to high in original array 
    indices_s = s[Num_zeros : Num_zeros + N_selectuion]

    #cordinates
    x = [int(i/w) for i in indices_s]  
    y = indices_s%w
    Cordinates = list(zip(x,y))

    temp_RTVTR = []
    temp_RC = []
    temp_WLC = []

    for c in Cordinates:
        for r in range(len(temp_clean['Q_Ratios'][c])):
            temp_RTVTR.append(temp_clean['Q_Ratios'][c][0][r])
        temp_RC.append(temp_clean['Q_RC'][c])
        temp_WLC.append(temp_clean['Q_WLC'][c])

    Gen_RTVTR.append(np.mean(temp_RTVTR))
    Gen_RC.append(np.mean(temp_RC))
    Gen_WLC.append(np.mean(temp_WLC))
```

```python
Live_df = pd.DataFrame(columns=['RTVTR', 'Ridge_Count', 'White_Line_Count'])
Gen_df = pd.DataFrame(columns=['RTVTR', 'Ridge_Count', 'White_Line_Count'])
```

```python
Gen_df['RTVTR'] = Gen_RTVTR
Gen_df['Ridge_Count'] = Gen_RC
Gen_df['White_Line_Count'] = Gen_WLC
Live_df['RTVTR' ] = Live_RTVTR
Live_df['Ridge_Count'] = Live_RC
Live_df['White_Line_Count'] = Live_WLC
```

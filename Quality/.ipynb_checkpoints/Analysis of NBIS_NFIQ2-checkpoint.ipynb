```python
import os 
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from numpy import genfromtxt 
import numpy as np

Floc = '/home/citer/Desktop/GANS/Generated_images_512_JPEG/Live_Minuteas/'
Floc2 = '/home/citer/Desktop/GANS/Generated_images_512_JPEG/Gen_Minuteas/'
```

```python
Train_pd = pd.read_pickle('/home/citer/Desktop/Fingerprint_synthesize/Quality/train_df.p')
```

```python
Train_pd.FilePath = Train_pd.FilePath.apply(lambda x: x.split('/')[-1])
```

```python
Train_pd.head()
```

```python
'''Reading the Minuteas for live samples'''

'''Minuteas'''
Names = ['MN','MX/MY','DIR','REL','TYP','FTYP','FN','NX1,Ny1,RC1','NX2,NY2,RC2',
        'NX3,NY3,RC3','NX4,NY4,RC4','NX5,NY5,RC5']

Rig_minuta = []
Bif_minutea = []
AVG_Rel = []
AVG_Rel_Rid = []
AVG_Rel_Bif = []
REL = []


for i in range(len(Train_pd)):
    file = Train_pd.iloc[i,3]
    mindata= os.path.join(Floc, file[:-4]+'.min')
    try:
        P = pd.read_csv(mindata, sep=':',header=1, names=Names, encoding='utf_8') # for when there is no min detected
        P = P.fillna('') #filling the na with empty string for easier processing
    except: 
        f = open(mindata, 'r')
        temp = f.readlines()
        num = temp[4].count(':')
        f.close()
        P = pd.read_csv(mindata, sep=':',header=1, names=Names[0:num+1], encoding='utf_8')
    temp_rig = P[P.TYP =='RIG ']
    temp_bif = P[P.TYP =='BIF ']
    Rig_minuta.append(temp_rig.shape[0])
    Bif_minutea.append(temp_bif.shape[0])
    AVG_Rel.append(P.REL.mean())
    AVG_Rel_Rid.append(temp_rig.REL.mean())
    AVG_Rel_Bif.append(temp_bif.REL.mean())
    REL.append(P.REL.tolist())

    
Train_pd['Rig_minuta'] = Rig_minuta
Train_pd['Bif_minutea'] = Bif_minutea
Train_pd['AVG_Rel'] = AVG_Rel
Train_pd['AVG_Rel_Rid'] = AVG_Rel_Rid
Train_pd['AVG_Rel_Bif'] = AVG_Rel_Bif
```

```python
'''Size of the live fingerprints'''
BS = 8  #block size
Live_Size = []

for i in range(len(Train_pd)):
    file = Train_pd.iloc[i,3]
    ''' Low Contrast Map'''       
    data = genfromtxt(os.path.join(Floc, file[:-4]+'.lcm'), delimiter='', dtype=np.uint8)
    LC_Map = np.repeat(np.repeat(data, BS, axis=1),  BS, axis=0)  #expanding the blocks in both dimensions
    Live_Size.append(int(160000-np.sum(LC_Map))) # 512*512 - size of finger- white is 1 black is 0. 
```

```python
plt.imshow(LC_Map, cmap='gray')
```

```python
Train_pd['Size'] = Live_Size
```

```python
Train_pd = Train_pd.drop(columns=['set_type_fold_0', 'set_type_fold_1', 'set_type_fold_2', 'set_type_fold_3', 'set_type_fold_4', 'set_type_fold_5', 'set_type_fold_6', 'set_type_fold_7', 'set_type_fold_8', 'set_type_fold_0'])
```

```python
'''Reading the Minuteas for Generated samples'''
Gen_files = os.listdir(Floc2)
Gen_pd = pd.DataFrame()

'''Minuteas'''
Names = ['MN','MX/MY','DIR','REL','TYP','FTYP','FN','NX1,Ny1,RC1','NX2,NY2,RC2',
        'NX3,NY3,RC3','NX4,NY4,RC4','NX5,NY5,RC5']

Files = []
Rig_minuta = [] # ridge minutiae 
Bif_minutea = [] # biforicaition minutiae
AVG_Rel = [] # average reliability
AVG_Rel_Rid = [] # average reliability in ridge minutiaes
AVG_Rel_Bif = [] # average reliability in bif minutiaes
REL = [] # reliability of all minutiaes


for file in Gen_files:
    if file[-3:] == 'min':
        Files.append(file)
        mindata= os.path.join(Floc2, file[:-4]+'.min')
        try:
            P = pd.read_csv(mindata, sep=':',header=1, names=Names, encoding='utf_8') # for when there is no min detected
            P = P.fillna('') #filling the na with empty string for easier processing
        except: 
            f = open(mindata, 'r')
            temp = f.readlines()
            num = temp[4].count(':')
            f.close()
            P = pd.read_csv(mindata, sep=':',header=1, names=Names[0:num+1], encoding='utf_8')
        temp_rig = P[P.TYP =='RIG ']
        temp_bif = P[P.TYP =='BIF ']
        Rig_minuta.append(temp_rig.shape[0])
        Bif_minutea.append(temp_bif.shape[0])
        AVG_Rel.append(P.REL.mean())
        AVG_Rel_Rid.append(temp_rig.REL.mean())
        AVG_Rel_Bif.append(temp_bif.REL.mean())
        REL.append(P.REL.tolist())

Gen_pd['Files'] = Files
Gen_pd['Rig_minuta'] = Rig_minuta
Gen_pd['Bif_minutea'] = Bif_minutea
Gen_pd['AVG_Rel'] = AVG_Rel
Gen_pd['AVG_Rel_Rid'] = AVG_Rel_Rid
Gen_pd['AVG_Rel_Bif'] = AVG_Rel_Bif
```

```python
'''Size of the generated fingerprints'''
BS = 8  #block size
Gen_files = sorted(os.listdir(Floc2))
Size = []
Files = []
for file in Gen_files:
    if file[-3:] == 'min':
        ''' Low Contrast Map'''       
        data = genfromtxt(os.path.join(Floc2, file[:-4]+'.lcm'), delimiter='', dtype=np.uint8)
        LC_Map = np.repeat(np.repeat(data, BS, axis=1),  BS, axis=0)  #expanding the blocks in both dimensions
        Size.append(int(262144 - np.sum(LC_Map))) # 512*512 - size of finger  - in pixels- estimate
        Files.append(file)
Gen_pd['Size'] = Size
```

```python
Gen_pd = Gen_pd.sort_values('Files')
Gen_pd = Gen_pd.reset_index()
```

```python
'''NFIQ2 Scores'''

Nloc = '/home/citer/Desktop/GANS/Generated_images_512_BMP/NFIQ2/'
Nloc2 = '/home/citer/Desktop/GANS/Generated_images_512_BMP/NFIQ2_Live/'

'''Size of the live fingerprints'''
Live_NFIQ2 = []

for i in range(len(Train_pd)):
    file = Train_pd.iloc[i,3]
    f = open(os.path.join(Nloc2, file[:-3]+'txt'))
    Scores = f.readlines()
    f.close()
    for line in Scores:
        if line[0:30]=='NFIQ2: Achieved quality score:':            
            Live_NFIQ2.append(line.split(':')[-1].split('\n')[0])

Train_pd['NFIQ2'] = Live_NFIQ2
```

```python
Train_pd['NFIQ2'] =Train_pd['NFIQ2'].astype(int)
```

```python
Gen_files = sorted(os.listdir('/home/citer/Desktop/GANS/Generated_images_512_BMP/Gen_512_BMP/gray/'))
Gen_NFIQ2 = []
Files = []

for file in Gen_files:
    f = open(os.path.join(Nloc, file[:-3]+'txt'))
    Scores = f.readlines()
    f.close()
    wrong = True
    for line in Scores:
        if line[0:30]=='NFIQ2: Achieved quality score:':            
            Gen_NFIQ2.append(line.split(':')[-1].split('\n')[0])
            wrong = False
    if wrong == True:
        Gen_NFIQ2.append(np.NAN)
    Files.append(file)
Gen_pd['NFIQ2'] = Gen_NFIQ2
```
